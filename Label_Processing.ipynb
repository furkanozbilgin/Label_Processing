```python
####  YOLO 1.1 (Darknet) formatı için 
####  NOT: Datasetin içinde çalıştırın ama fazla miktar data olunca çökebiliyor. :/
import numpy as np
import os

### Girilen label numarasındaki etiketleri siler
def Remove_Label(label):

    #print(len(os.listdir()))    ## number of files
    filename = os.listdir()     ## filename
    
    for file in filename:
        #print(file)
        if file.endswith(".txt"): 
            with open(file, "r") as input:
                with open("temp.txt", "w") as output:
                    # iterate all lines from file
                    for line in input:
                        # if line starts with substring 'time' then don't write it in temp file
                        if not line.strip("\n").startswith(label):
                            output.write(line)

            # replace file with original name
            os.replace('temp.txt', file)

            
### Boş txt ve ona karşılık gelen imageleri siler          
def Delete_Unannotated():
    
    #print(len(os.listdir()))                    ## number of files
    filename = os.listdir()                     ## filename
    sayi = 0
    sayi1 = 0
    sayi2 = 0
    for file in filename:
        #print(file)
        if file.endswith(".txt"):  
            if not os.path.getsize(file):                ## boş txt dosyları tespit edilir
                sayi += 1
                print((file))
                os.remove(file)                          ## tespit edilen txt dosyalarını siler

                replace_format_to_jpg = file.replace(".txt", ".jpg")             # txt convert to jpg
                replace_format_to_png = replace_format_to_jpg.replace(".jpg", ".png")     # txt convert to png 


                if replace_format_to_jpg in filename:
                    sayi1 += 1
                    print(replace_format_to_jpg)
                    os.remove(replace_format_to_jpg)
                    
                elif replace_format_to_png in filename:
                    sayi2 += 1
                    print(replace_format_to_jpg)
                    os.remove(replace_format_to_png)
                   
                else:
                    print("Uyarı!" ,file,"dosyasına karsilik gelen image yoktur")
    print("Silinen txt sayisi:",(sayi))
    sayi3 = sayi1+ sayi2
    print("Silinen image sayisi:",(sayi3))
```

```python
label = input("enter a label number:")

Remove_Label(label)
Delete_Unannotated()
```
